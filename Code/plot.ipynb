```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys

sys.path.append("../")
```

```python
data1 = pd.read_csv(r'../Results/BINDINGDB_18/results.txt', sep='\t', nrows=20, header=None)
data2 = pd.read_csv(r'../Results/BIOSNAP_18/results.txt', sep='\t', nrows=20, header=None)
data3 = pd.read_csv(r'../Results/DRUGBANK_18/results.txt', sep='\t', nrows=20, header=None)
data4 = pd.read_csv(r'../Results/DAVIS_18/results.txt', sep='\t', nrows=20, header=None)
data5 = pd.read_csv(r'../Results/E_18/results.txt', sep='\t', nrows=20, header=None)
data6 = pd.read_csv(r'../Results/GPCR_18/results.txt', sep='\t', nrows=20, header=None)
data7 = pd.read_csv(r'../Results/IC_18/results.txt', sep='\t', nrows=20, header=None)
data8 = pd.read_csv(r'../Results/NR_18/results.txt', sep='\t', nrows=20, header=None)
```

```python
auc1 = pd.DataFrame({'AUC': data1.iloc[:, 0], 'Database': 'BINDINGDB'})
auc2 = pd.DataFrame({'AUC': data2.iloc[:, 0], 'Database': 'BIOSNAP'})
auc3 = pd.DataFrame({'AUC': data3.iloc[:, 0], 'Database': 'DRUGBANK'})
auc4 = pd.DataFrame({'AUC': data4.iloc[:, 0], 'Database': 'DAVIS'})
auc5 = pd.DataFrame({'AUC': data5.iloc[:, 0], 'Database': 'E'})
auc6 = pd.DataFrame({'AUC': data6.iloc[:, 0], 'Database': 'GPCR'})
auc7 = pd.DataFrame({'AUC': data7.iloc[:, 0], 'Database': 'IC'})
auc8 = pd.DataFrame({'AUC': data8.iloc[:, 0], 'Database': 'NR'})
all_data = pd.concat([auc1, auc2, auc3])
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Database', y='AUC', data=all_data, inner='quartile', palette='muted')
plt.title('AUC Distribution Across Databases')
plt.xlabel('Database')
plt.ylabel('AUC')
plt.savefig('../Results/violin_plot_AUC.pdf', format='pdf')
plt.show()
```

```python
import os
# AUC_heatmap
databases = ['DrugBank', 'BIOSNAP', 'BindingDB', 'Davis', 'E', 'GPCR', 'IC', 'NR']
data=np.zeros((8,8))
folder_path=r'../Results/'
for i, db1 in enumerate(databases):
    for j, db2 in enumerate(databases):
        if i != j: 
            file_name = f'{db1}_WO_{db2}_18/results.txt'  
            file_path = os.path.join(folder_path, file_name)
            
            
            if os.path.exists(file_path):
                df = pd.read_csv(file_path,sep='\t',nrows=5, header=None)
                auc_value = df.iloc[:5].mean().values[0]  
                data[j, i] = auc_value
            else:
                print(f"Warning: {file_path} does not exist.")
        else:
            data[i, j] = np.nan
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# plot_AUC_
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".4f", cmap="YlGnBu", cbar=True,
            xticklabels=["DrugBank", "BIOSNAP", "BindingDB", "Davis", "E", "GPCR", "IC", "NR"],
            yticklabels=["DrugBank", "BIOSNAP", "BindingDB", "Davis", "E", "GPCR", "IC", "NR"])
plt.title('AUROC', fontsize=15, fontweight='bold')
plt.ylabel("Evaluated",fontweight='bold',fontsize=15)
plt.xlabel("Trained",fontweight='bold',fontsize=15)
plt.savefig('../Results/C_AUC.svg', format='svg')
plt.show()
```

```python
#AUPR_heatmap
databases = ['DrugBank', 'BIOSNAP', 'BindingDB', 'Davis', 'E', 'GPCR', 'IC', 'NR']
data=np.zeros((8,8))
folder_path=r'../Results/'
for i, db1 in enumerate(databases):
    for j, db2 in enumerate(databases):
        if i != j: 
            file_name = f'{db1}_WO_{db2}_18/results.txt'  
            file_path = os.path.join(folder_path, file_name)
            
            
            if os.path.exists(file_path):
                df = pd.read_csv(file_path,sep='\t',usecols=[1],nrows=5, header=None)
                aupr_value = df.mean().values[0]  
                data[j, i] = aupr_value
            else:
                print(f"Warning: {file_path} does not exist.")
        else:
            data[i, j] = np.nan
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".4f", cmap="YlGnBu", cbar=True,
            xticklabels=["DrugBank", "BIOSNAP", "BindingDB", "Davis", "E", "GPCR", "IC", "NR"],
            yticklabels=["DrugBank", "BIOSNAP", "BindingDB", "Davis", "E", "GPCR", "IC", "NR"])
plt.title('AUPRC', fontsize=15, fontweight='bold')
plt.ylabel("Evaluated",fontweight='bold',fontsize=15)
plt.xlabel("Trained",fontweight='bold',fontsize=15)
plt.savefig('../Results/C_AUPR.svg', format='svg')
plt.show()
```
